```python
df.Cinsiyet.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0,1])],remainder='passthrough')
ct.fit_transform(df)
encoded = pd.DataFrame(ct.fit_transform(df))
encoded
```

```python
encoded.drop([6],axis=1,inplace=True)
encoded.head()
```

```python
encoded.columns= list(range(len(encoded.columns)))
encoded.head()
```

```python
encoded.to_csv("/workspace/KordsaIntroML/Documents/DataPreprocessing/data/sampleV2.csv",index=False)
```
