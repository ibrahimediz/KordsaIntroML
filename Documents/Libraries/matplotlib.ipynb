```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x**2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.hist(np.random.random(1000));
ax4.hist(np.random.randn(1000));
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x**2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].hist(np.random.random(1000));
ax[1,1].hist(np.random.randn(1000));
```

```python
dfKalp = pd.read_csv("/workspace/KordsaIntroML/Datasets/heart.csv")
dfKalp.head()
```

```python
dfKalp.plot.bar()
```

```python
dfKalp.plot.hist(figsize=(10,30),subplots=True,bins=50);
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,8))

scatter = ax0.scatter(dfKalp["AGE_50"],dfKalp["CHOL_50"],c=dfKalp["DEATH"])
ax0.set(title="Kalp Rahatsızlığı ve Kolesterol Seviyeleri",ylabel="Kolesterol");
ax0.legend(*scatter.legend_elements(),title="DEATH");
ax0.set_xlim([50,80])
ax0.axhline(dfKalp["CHOL_50"].mean(),linestyle="--",color="b");

scatter = ax1.scatter(dfKalp["AGE_50"],dfKalp["HT_50"],c=dfKalp["DEATH"])
ax1.set(title="Kalp Rahatsızlığı ve Maksimum Kalp Atışı Seviyeleri",xlabel="Yaş",ylabel="Maksimum Kalp Atışı");
ax1.legend(*scatter.legend_elements(),title="DEATH");
ax1.axhline(dfKalp["HT_50"].mean(),linestyle="--",color="r");
ax1.set_ylim([60,200])
fig.suptitle("Kalp Rahatsızlığı Analizi");
```
