### DataFrame

```python
liste2 = ["a","b","c","d"]
seri2 = pd.Series(liste2)
liste = [1,2,3,4]
seri1 = pd.Series(liste)
df = pd.DataFrame({"liste1":seri1,"liste2":seri2})
df
```

```python
liste2 = ["a","b","c","d"]
seri2 = pd.Series(liste2)
liste1 = [1,2,3,4]
seri1 = pd.Series(liste1)
liste3 = [10,20,30,40]
seri3 = pd.Series(liste3,index=["a","b","c","d"])
df = pd.DataFrame({"liste1":seri1,"liste2":seri2,"liste3":seri3})
df
```

#### Group BY

```python
kalp.groupby("Disease").age.max()
```

```python
kalp.groupby("Disease").age.min()
```

```python
kalpista = kalp.groupby(["age","sex"]).Disease.agg([len,sum])
```

```python
kalpista.index
```

```python
kalpista = kalpista.reset_index()
```

```python
kalpista
```
